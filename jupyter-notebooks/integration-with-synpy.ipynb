# Интегрирование функций одной переменной

```python
from sympy import *
init_printing()

# визуализация интеграла
import numpy as np
from matplotlib.patches import Polygon
import matplotlib.pyplot as plt
```

```python
x = symbols('x')
```

# Неопределённый интеграл

$$
	F(x)=\int f(x)dx
$$

зададим функцию $f$

```python
f_expr = x*cos(x)
f_expr
```

её первообразная

```python
integrate(f_expr, x)
```

```python
# альтернативно
f_expr.integrate(x)
```

значение первообразной в точке $x=\pi$

```python
integrate(f_expr, (x, pi))
```

## Определённый интеграл

$$
	\int_a^b f(x)dx
$$

зададим функцию $f$

```python
f_expr = x*exp(x)
f_expr
```

вычислим $\int_0^1 f(x)dx$

```python
integrate(f_expr, (x, 0, 1))
```

```python
#  альтернативно
f_expr.integrate((x,0,1))
```

## Геометрическая интерпретация интеграла 

```python
def f(x):
	return np.exp(-0.1*x**2)

x_min, x_max = -7, 7
x_grid = np.linspace(x_min, x_max, 100)

a, b = -4, 3

fig, ax = plt.subplots(figsize=(12, 6))

ax.plot(x_grid, f(x_grid), 'r')

ix = np.linspace(a, b, 100)
iy = f(ix)

verts = [(a, 0), *zip(ix, iy), (b, 0)]
poly = Polygon(verts, facecolor='0.9', edgecolor='0.5')
ax.add_patch(poly)
ax.text(0, 0.3, r"$P(a\leq X\leq b)=\int_a^b f(t)dt$", horizontalalignment='center', fontsize=15)

ax.hlines(0, x_min, x_max, linestyles='dotted')
ax.text(a, 0, f"a", horizontalalignment='center', verticalalignment='top', fontsize=11)
ax.text(b, 0, f"b", horizontalalignment='center', verticalalignment='top', fontsize=11)

plt.show()
```
